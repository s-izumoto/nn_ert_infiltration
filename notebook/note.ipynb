```python

import importlib, sys
import importlib.metadata as md
import numpy as np

print("NumPy:", np.__version__)
candidates = [
    # よく落ちる順に並べています
    "pygimli","scipy","numba","sklearn","pandas","matplotlib",
    "scikit_image","shapely","rasterio","xgboost","torch","opencv_python"
]

def dist_of(mod):
    try:
        # top-level→distribution の逆引き
        mapping = md.packages_distributions()
        dists = mapping.get(mod.replace('.', '-'), []) or mapping.get(mod, [])
        return ",".join(dists) if dists else "?"
    except Exception:
        return "?"

for m in candidates:
    try:
        importlib.import_module(m)
        print(f"[OK]   {m}  (dist={dist_of(m)})")
    except Exception as e:
        print(f"[FAIL] {m}  (dist={dist_of(m)}) -> {e.__class__.__name__}: {e}")

```

```python
import numpy as np
import pandas as pd

npz_path = "mean_values.npz"  # ←必要に応じて変更

def small_preview(a, max_items=6):
    """配列の冒頭だけ文字列でプレビュー"""
    if a.ndim == 0:
        # スカラー
        try:
            return repr(a.item())
        except Exception:
            return str(a)
    if a.ndim == 1:
        return np.array2string(a[:max_items], precision=4, threshold=max_items)
    if a.ndim == 2:
        r = min(3, a.shape[0])
        c = min(6, a.shape[1])
        return np.array2string(a[:r, :c], precision=4, threshold=r*c)
    # 3次元以上は先頭スライスを表示
    head = a[(0,) * (a.ndim - 2)]
    r = min(3, head.shape[0])
    c = min(6, head.shape[1])
    return f"slice[0,...]:\n" + np.array2string(head[:r, :c], precision=4, threshold=r*c)

with np.load(npz_path, allow_pickle=True) as z:
    rows = []
    for k in z.files:
        a = z[k]
        size_mb = round(a.nbytes / (1024**2), 3)
        rows.append((k, a.shape, str(a.dtype), size_mb, small_preview(a)))

df = pd.DataFrame(rows, columns=["key", "shape", "dtype", "size_MB", "preview"]).sort_values("key")
df

```
